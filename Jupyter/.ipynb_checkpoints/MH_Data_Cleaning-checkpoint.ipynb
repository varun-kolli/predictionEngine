# Creating a balanced dataset for training

```python
# for reproducibility purposes
seed = 100
# SMOTE number of neighbors
k = 75

df_balanced = df_relevant.copy()
# make a new df made of all the columns, except the target class
x = pd.get_dummies(df_balanced.drop(columns = 'MH1'), drop_first = True)
y = df_balanced['MH1']
sm = SMOTE(sampling_strategy = 'auto', k_neighbors = k, random_state = seed)
x_res, y_res = sm.fit_resample(x, y)

df_balanced = pd.concat([pd.DataFrame(x_res), pd.DataFrame(y_res)], axis = 1)
df_balanced.shape
```

# Model with no null values

```python
df_nonulls = df_relevant.copy()
x = pd.get_dummies(df_nonulls.drop(columns = 'MH1'), drop_first = True)
y = df_nonulls['MH1']
df_nonulls = pd.concat([pd.DataFrame(x), pd.DataFrame(y)], axis = 1)
df_nonulls = df_nonulls.dropna()

df_balanced_nonulls = df_balanced.dropna()
print(df_nonulls.shape)
print(df_balanced_nonulls.shape)
```

```python
#split into train and test
# train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.4, random_state = 1) 

train_x = df_balanced_nonulls.drop(columns = 'MH1')
train_y = df_balanced_nonulls['MH1']
test_x = pd.get_dummies(df_nonulls.drop(columns = 'MH1'), drop_first = True)
test_y = df_nonulls['MH1']

#defining the folds
folds = KFold(n_splits = 5, shuffle = True, random_state = 1)

clf = DecisionTreeClassifier(random_state = 42)

hyper_params = {
    'max_depth': [10, 60 , 100], 
    'min_samples_split': [10, 20, 40], 
    'min_samples_leaf': [45, 55, 65]
}

# Call GridSearchCV()
model_clf = GridSearchCV(estimator = clf, 
                        param_grid = hyper_params, 
                        scoring = 'f1_weighted', 
                        cv = folds, 
                        verbose = 1,
                        n_jobs = -1) # Will utilize all available CPUs 

model_clf.fit(train_x, train_y)  
print('Initial score: ', model_clf.best_score_)
print('Initial parameters: ', model_clf.best_params_)
```

```python
hyper_params_new = {
    'max_depth': list(range(20, 24)), 
    'min_samples_split': list(range(1, 5)), 
    'min_samples_leaf': list(range(42, 47))
}

# Call GridSearchCV()
model_clf = GridSearchCV(estimator = clf, 
                        param_grid = hyper_params_new, 
                        scoring = 'f1_weighted', # Use a suitable regression metric
                        cv = folds, 
                        verbose = 1,
                        n_jobs = -1) # Will utilize all available CPUs 

model_clf.fit(train_x, train_y)  
print('Improved score: ', model_clf.best_score_)
print('Improved parameters: ', model_clf.best_params_)
```

```python
bestTree = model_clf.best_estimator_
```

```python
# Displaying feature importance
importance = model_clf.best_estimator_.feature_importances_
feature_imp = pd.DataFrame(list(zip(train_x.columns, importance)),
               columns = ['Feature', 'Importance'])

feature_imp = feature_imp.sort_values('Importance', ascending = False).reset_index(drop = True)

feature_imp.head()
```

```python
# Specifying figure size
fig, ax = plt.subplots(figsize=(6, 6)) 

# Generating confusion matrix
plot_confusion_matrix(bestTree, test_x, test_y,
                      cmap = plt.cm.Blues,
                      values_format = '',
                      ax = ax);
```

```python
# Predictions on test set
y_pred = bestTree.predict(test_x)

# Generating Classification Report
print("Classification Report - \n", 
      classification_report(test_y, y_pred))
```

# Model with nulls replace with distribution

```python
df_dist = df_relevant.copy()
df_dist
```

```python
# finding the distribution

total_len = len(df_dist)
dist_ser = df_dist['AGE'].value_counts()
dist_df = dist_ser.to_frame()
dist_df['cat'] = dist_df.index.tolist()
numb_lst = list(range(0, len(dist_df.index.tolist())))
dist_df.set_index([pd.Index(numb_lst)], inplace = True)
prop_lst = []

# make proportions of nulls in each column a dataframe
for numb in dist_df['AGE']:
    prop = numb / total_len
    prop_lst.append(prop)
    
dist_df['proportion'] = prop_lst

# counting the number of nulls
null_filt = df_dist['AGE'].isna()
len_nulls = len(df_dist[null_filt])
print(len_nulls)
index = 0

def repeater(n, row, age):
    for x in range(int(n)):
        if pd.isna(row['AGE']):
            ret = age
        else:
            ret = row['AGE']
        return ret

for age in dist_df['cat']:
    print(age)
    changes = len_nulls * prop_lst[index]
    print(changes)
    df_dist['AGE'] = df_dist.apply(lambda row: repeater(changes, row, age), axis = 1)
    index = index + 1
```

```python
def dist(column):
    total_len = len(df_dist)
    dist_ser = df_dist[column].value_counts()
    dist_df = dist_ser.to_frame()
    dist_df['cat'] = dist_df.index.tolist()
    numb_lst = list(range(0, len(dist_df.index.tolist())))
    dist_df.set_index([pd.Index(numb_lst)], inplace = True)
    prop_lst = []

    # make proportions of nulls in each column a dataframe
    for numb in dist_df[column]:
        prop = numb / total_len
        prop_lst.append(prop)

    dist_df['proportion'] = prop_lst

    # counting the number of nulls
    null_filt = df_dist[column].isna()
    len_nulls = len(df_dist[null_filt])
    index = 0

    def repeater(n, row, item):
        for x in range(int(n)):
            if pd.isna(row[column]):
                ret = item
            else:
                ret = row[column]
            return ret

    for item in dist_df['cat']:
        changes = len_nulls * prop_lst[index]
        df_dist[column] = df_dist.apply(lambda row: repeater(changes, row, item), axis = 1)
        index = index + 1
```
